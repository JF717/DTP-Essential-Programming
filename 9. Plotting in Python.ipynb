We can use the ylabel and xlabel functions to put the labels into the plots.

```python
x = np.linspace(0,8 * np.pi,500)
y = np.sin(x)
plt.plot(x,y)
plt.title("A sin Wave");
```

How about we try plotting some of the data we made earlier with our financial functions. 

```python
def savings(start,interest,time):
    interest = 1 + (interest/100)
    current = []
    current.append(start)
    for i in range(0,time):
        current.append(current[i] * interest)
    
    return(current)
```

```python
sav = savings(1000,4,20)
```

```python
plt.plot(sav)
plt.ylabel("Investment Value £")
plt.xlabel("Year")
plt.title("Return on Investment")
```

That's pretty good plotting right there. Your money just keeps growing. Try plotting it for 100 years rather than 20

Lets plot some data from the iris dataset that we imported earlier.

```python
Data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

```python
plt.plot(Data["sepal_length"],Data["petal_length"])
```

Well that is totally useless isn't it. I guess we might have to get a bit more complicated in our plotting.

```python
plt.scatter(Data["sepal_length"],Data["petal_length"])
```

Ok that is much better isn't it. There are multiple different kinds of plot in matplotlib that we can use to visualise things better.

How about we mark them in different colours based on the species. This uses some of our techniques from the pandas session. Sorting data by the species and then using the c = "" command. c is colour and we can put either the full color name in it or lots of colours work shortened like b is blue, r is red etc. 

```python
plt.scatter(Data.sepal_length[Data["species"]=="setosa"],Data.petal_length[Data["species"]=="setosa"],c="r")
plt.scatter(Data.sepal_length[Data["species"]=="virginica"],Data.petal_length[Data["species"]=="virginica"],c="b")
plt.scatter(Data.sepal_length[Data["species"]=="versicolor"],Data.petal_length[Data["species"]=="versicolor"],c="g")

plt.xlabel("Petal Length")
plt.ylabel("Sepal Length")
plt.title("Flower Sepal Against Petal Length")
plt.legend(["Setosa","Virginica","Versicolor"])
```

We can plot multiple things at the same time using the subplot functions. With subplots we have to call the subplots command and specify how many we want. This function returns two things the fig which is the overall object and the ax which is an array of the two plots. We alter the two plots by calling commands in the ax.

```python
fig,ax = plt.subplots(1,2,figsize=(20,10))
ax[0].scatter(Data.sepal_length[Data["species"]=="setosa"],Data.petal_length[Data["species"]=="setosa"],c="r")
ax[0].scatter(Data.sepal_length[Data["species"]=="virginica"],Data.petal_length[Data["species"]=="virginica"],c="b")
ax[0].scatter(Data.sepal_length[Data["species"]=="versicolor"],Data.petal_length[Data["species"]=="versicolor"],c="g")

ax[1].scatter(Data.sepal_width[Data["species"]=="setosa"],Data.petal_width[Data["species"]=="setosa"],c="r")
ax[1].scatter(Data.sepal_width[Data["species"]=="virginica"],Data.petal_width[Data["species"]=="virginica"],c="b")
ax[1].scatter(Data.sepal_width[Data["species"]=="versicolor"],Data.petal_width[Data["species"]=="versicolor"],c="g")

ax[0].set_xlabel("Sepal Length")
ax[0].set_ylabel("Petal Length")
ax[1].set_xlabel("Sepal Width")
ax[1].set_ylabel("Petal Width")

ax[0].legend(["Setosa","Virginica","Versicolor"])
ax[1].legend(["Setosa","Virginica","Versicolor"])

ax[1].set_title("Flower Width")
ax[0].set_title("Flower Length")
```
